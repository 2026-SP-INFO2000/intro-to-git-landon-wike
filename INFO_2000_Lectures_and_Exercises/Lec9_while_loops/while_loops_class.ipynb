# While Loops

**Learning goals:**
- Understand the syntax and logic of a `while` loop
- Know how it differs from a `for` loop
- Use conditions to control repetition
- Apply `break` and `continue` to control flow
- Avoid infinite loops


## 0) Motivation 
- *What if we don’t know how many times a loop should run?*
- Example: asking for input until the user types `'quit'`
- Contrast:
  - `for` = repeat a **known** number of times
  - `while` = repeat **until a condition becomes false**

## 1) Basic Syntax 
```
while CONDITION:
    BODY
```

````python
count = 0
while count < 5:
    print("Count is", count)
    count += 1
````

## 2) Infinite Loops
⚠️ Be careful with conditions that are always True.

````python
# while True:
#     print("This runs forever!")
````

## 3) Using Input with `while True`

````python
word = ""
while word != "quit":
    word = input("Type a word (or 'quit' to stop): ")
    print("You typed:", word)
````

## 4) Using `break` and `continue` in `while True`

````python
# break example
n = 0
while n < 10:
    if n == 5:
        break
    print(n)
    n += 1
````

````python
# continue example
n = 0
while n < 10:
    n += 1
    if n % 2 == 0:
        continue
    print(n)   # prints only odd numbers
````

## 5) While vs For Loops 
- `for` = iterate over a sequence (lists, strings, ranges)
- `while` = repeat until a condition fails
- Sometimes interchangeable, but use whichever matches the problem

````python
n = 5
while n > 0:
    print(n)
    n -= 1
````

## Example 1 — Guess the Number 🎯
Generate a random number between 1 and 10.
Keep asking the user to guess until they’re correct.
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"

````python
import random
target = random.randint(1, 10)   # random int between 1 and 10 (inclusive)

while True:
    guess = int(input("Please input your guess:"))

    if guess == target:
        print("Correct!")
        break
    elif guess == target - 1 or guess == target + 1:
        print(f"{guess} is close!")
    elif guess > target+1:
        print(f"{guess} is too High!")
    else:
        print(f"{guess} is too Low!")
````

## Example 2 — Password Prompt 🔑
Keep asking for a password until the user enters `'python123'`.
If correct, print `'Access granted!'`. 

````python
while True:
    pwd = input("Please input your password:")
    if pwd == 'python123':
        print('Access granted!')
        break
````

## Example 3 — Sum Until Zero ➕
Ask the user to enter numbers. Keep summing until the user enters `0`. Print the total.

````python
sum_res = 0
while True:
    num = int(input("Please input a number:"))
    if num == 0:
        break
    sum_res += num

print(sum_res)
````

## Example 4 — Collatz Conjecture
The Collatz Conjecture is a famous unsolved problem in mathematics. The rules are simple:
1. Start with any positive integer n.

2. Repeat the following steps until n becomes 1:

- If n is even, divide it by 2.
- If n is odd, multiply it by 3 and add 1.

````python
n = int(input("Enter a positive integer: "))

while n != 1:
    print(n, end=" -> ")
    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1

print(1)
````

## Example 5 ATM Menu 💳
Simulate a simple ATM menu:
- Options: `'balance'`, `'deposit'`, `'withdraw'`, `'Exit'`
- Use a while loop until the user chooses quit.

````python
balance = 1000
while True:
    print("\nATM Menu:")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print(f"Balance: ${balance}")
    elif choice == "2":
        amount = float(input("Enter deposit amount: "))
        balance += amount
        print("Deposited successfully.")
    elif choice == "3":
        amount = float(input("Enter withdrawal amount: "))
        if amount <= balance:
            balance -= amount
            print("Withdrawn successfully.")
        else:
            print("Insufficient funds!")
    elif choice == "4":
        print("Thank you for using the ATM.")
        break
    else:
        print("Invalid choice.")
````

## Wrap‑Up
- **While loops** are for conditions that change over time
- Always ensure the condition eventually becomes False → avoid infinite loops
- Use `break`/`continue` to fine-tune loop behavior
- For known repetitions, prefer `for`; for unknown, use `while`
