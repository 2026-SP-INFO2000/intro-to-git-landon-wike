### Exercise 1
Keep asking for a password until the user enters `'python123'` or the maximum number of attempts is reached (You can use the input function to input the maximum number of attempts).
If correct, print `'Access granted!'`. 

```python
max_attempts = int(input('The maximum number of attempts are: '))
attempts = 0 
attempts_left = max_attempts - attempts
while attempts < max_attempts:
    pwd_guess = input('Input password: ')
    if pwd_guess == 'python123':
        print('Access Granted!')
        break
    else:
        attempts += 1
        print('Incorrect password.')
if attempts == max_attempts:
    print('You have reached the maximum attempts. Access denied.')
```

### Exercise 2 Write the number guessing game using while loops. 
Keep guessing until correct or the number of guesses is exhausted.
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"


```python
import random
target = random.randint(1, 10)
max_guesses = 10
count = 0
while count < max_guesses:
    count += 1
    guess = int(input('Input your guess:'))

    if guess == target:
        print('Correct!')
        break
    elif guess == target - 1 or guess == target + 1:
        print('Close!')
    elif guess > target+1:
        print('Too High!')
    else:
        print('Too Low!')
if count == max_guesses:
    print('You are out of guesses.')
```

### Exercise 3 Cookie Monster Game

Start with a number of cookies (use input function). Each loop, the “monster” eats 1–3 cookies (random). Keep looping until no cookies remain. Print how many bites it took.
- `import random` and `cookies_per_bite = random.randint(1, 3)`

```python
import random
cookies = int(input('Enter amount of cookies: '))
count = 0
while cookies > 0:
    cookies_per_bite = random.randint(1, 3)


    if cookies_per_bite > cookies:
        cookies_per_bite = cookies
    
    cookies -= cookies_per_bite
    count += 1
    
print('It took', count,'bites to finish the cookies')  
```

### Exercise 4
Design your own ATM menu, with basic operations `'balance'`, `'deposit'`, `'withdraw'`, `'Exit'`.

```python
#### balance = 4000
while True:
    print('\nATM Menu:')
    print('1. Balance')
    print('2. Deposit')
    print('3. Withdraw')
    print('4. Exit')

    choice = input('Enter your choice: ')

    if choice == '1':
        print(f"Balance: ${balance}")
    elif choice == '2':
        amount = float(input('Enter deposit amount: '))
        balance += amount
        print('Deposited successfully. Current balance is', balance)
    elif choice == "3":
        amount = float(input('Enter withdrawal amount: '))
        if amount <= balance:
            balance -= amount
            remaining_balance = balance
            print('Withdrawn successfully. Balance remaining is', remaining_balance)
        else:
            print('Insufficient funds!')
    elif choice == "4":
        print('Thank you for using the ATM.')
        break
    else:
        print('Invalid choice.')
```
