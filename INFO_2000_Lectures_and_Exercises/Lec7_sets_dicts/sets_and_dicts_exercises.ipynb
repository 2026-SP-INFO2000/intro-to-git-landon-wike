### Exercise 1 — Unique Letters Finder ✍️
Write a program that asks the user for a word and prints all the **unique letters** in it.
```python
# Example input: "banana"
# Expected output: {'b', 'a', 'n'}
```

````python
word = input("Enter a word: ")

unique_letters = set(word)

print(unique_letters)
````

### Exercise 2 — Students in Multiple Clubs 🎶⚽
You are given two sets:
```python
music_club = {"Alice", "Bob", "Charlie"}
sports_club = {"Bob", "David", "Charlie"}
```
- Find students who are in **both clubs**.
- Find students who are in **either club but not both**.

````python
music_club = {"Alice", "Bob", "Charlie"}
sports_club = {"Bob", "David", "Charlie"}


both_clubs = music_club & sports_club
print("In both clubs:", both_clubs)


either_not_both = music_club ^ sports_club
print("In either club but not both:", either_not_both)
````

### Exercise 3 — Character Counter 🔤
Write a program that counts how many times each character appears in a string using a dictionary.
```python
# Input: "hello"
# Expected: {'h':1, 'e':1, 'l':2, 'o':1}
```

````python

text = input("Enter a string: ")

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
````

### Exercise 4 — Inventory System 🛒
You have a dictionary representing a store’s inventory:
```python
inventory = {"apple": 5, "banana": 2, "orange": 0}
```
- Ask the user what they want to buy.
- If the item is available (quantity > 0), reduce the count by 1 and confirm purchase.
- If not, print `'Sorry, out of stock'`.
- Stop when the user types `'quit'`. 

````python
inventory = {"apple": 5, "banana": 2, "orange": 0}

while True:
    item = input("What would you like to buy? (type 'quit' to stop): ").lower()
    
    if item == "quit":
        print("Thank you for shopping!")
        break
    
    if item in inventory:
        if inventory[item] > 0:
            inventory[item] -= 1
            print(f"Purchase successful! {item} remaining: {inventory[item]}")
        else:
            print("Sorry, out of stock.")
    else:
        print("Item not found.")

print("Final inventory:", inventory)
````
