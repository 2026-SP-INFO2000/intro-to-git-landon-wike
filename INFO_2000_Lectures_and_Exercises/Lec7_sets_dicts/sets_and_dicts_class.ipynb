## 1) Sets Basics

- Similar to the concept of a mathematical set, as an **unordered** collection of **unique** objects
- Useful to enforce uniqueness, and to use set operations
- Like tuples, immutable and not indexed (not subscriptable)
- Created as a comma separated list of objects enclosed by curly braces {}
{obj1,obj2,obj3,…,obj4}
- If any of the values are equivalent, only 1 of them will be included!


```python
s1 = {1, 2, (3)}
s2 = set([2, 3, 4])
print(s1)
print(type(s1))
```

```python
s = {1, 2, 2, 3}
print(s)  # duplicates removed
print(2 in s)  # membership test
```

**Q:** What can be the elements in a set?
**A**: Hashable objects like numbers, strings, and tuples; not lists, sets, and dictionaries

```python
a = {1,2,(2)}
a
```

### Common Operations
- `len(SET1) -> int`
Number of elements in the set
- `SET1.add(element)`: Add an element to a set
- `SET1.union(SET2) -> set`
Combines the sets; Can also use `SET1 | SET2`
- `SET1.difference(SET2) -> set`
Result is the set of elements that are in SET1 but not in SET2;
Can also use `SET1 - SET2` 
- `SET1.intersection(SET2) -> set`
Result only contains elements in both of the sets;
Can also use `SET1 & SET2`
- `SET1.symmetric_difference(SET2) - > set`
`(SET1-SET2).union(SET2-SET1)`
- `list(set(LIST)) -> list`
Useful for eliminating duplicate elements of LIST


```python
a = {1,2,3}
b = {3,4,5}
print(a | b)  # union
print(a & b)  # intersection
print(a - b)  # difference
print(a ^ b)  # symmetric difference
```

```python
lst = [1,2,2,3,3,4,5]
lst = list(set(lst))
print(lst)
```

```python

a = {1,2,"1"}
a.add(4)
a
```

### Example 1: Lucky Draw 🎲

```python
import random

participants = ["Alice", "Bob", "Charlie", "David", "Emma", "Frank"]
winners = set()

while len(winners) < 3:
    winners.add(random.choice(participants))

print("Winners:", winners)
```

### Example 2: Vowel Checker

```python
word = "education"
vowels = set("aeiou")

if vowels.issubset(set(word)):
    print("This word has all vowels!")
else:
    print("Missing some vowels")
```

### Example 3: Find Common Friends 👭

```python
friends_A = {"Alice", "Bob", "Charlie"}
friends_B = {"Bob", "David", "Emma"}

common = friends_A & friends_B
print("Common friends:", common)
```

## 2) Dictionaries Basics
- Most complicated container type
- Collection of key-value pairs
- Useful as a mutable collection of data that is sortable, and hence quickly retrievable, by the key
- Like sets, **keys** must be **unique**
- Created as a comma separated list of colon separated key:value pairs enclosed by curly braces

`{key1:value1,key2:value2,key3:value3,…,keyK:valueK}`

- Note that keys should be hashable, and tuples/strings and integers are most commonly used for keys


```python
student = {"name": "Alice", "age": 21, "major": "CS"}
print(student["name"])
```

### Common Operations
- `len(DICT) -> int`: Number of key:value pairs
- `DICT[KEY]` -> VALUE: Retrieves the VALUE corresponding to KEY in DICT
- `DICT.keys() -> dict_keys`: Retrieves only the keys of DICT as an ordered iterable
- `DICT.values() -> dict_values`: Retrieves only the values of DICT as an iterable, ordered on the keys
- `DICT.items() -> dict_items`: Retrieves the values and keys as an **ordered** list of **tuples (key,value)**


```python
student["age"] = 22
student["gpa"] = 3.9
del student["major"]
print(student)

for key, value in student.items():
    print(key, ":", value)
```

### Mutability of dictionaries
- `DICT[KEY] = OBJECT`
Change the object paired with KEY in DICT to OBJECT 
If KEY does not yet exist in DICT, creates the KEY

- `del DICT[KEY]`
Deletes the KEY and paired value from the dictionary


### Example 4: Word Counter 📖

```python
text = "python is fun and python is powerful"
text.split()
```

```python
text = "python is fun and python is powerful"
counts = {}

for word in text.split():
    if word in counts:
        counts[word] += 1
    else:
        counts[word] = 1

print(counts)
```

### Example 5: Simple Login System 🔑

```python
users = {"alice": "1234", "bob": "abcd"}
attempts = 0

while attempts < 3:
    name = input("Enter username: ")
    pwd = input("Enter password: ")
    if name in users and users[name] == pwd:
        print("Login successful!")
        break
    else:
        print("Invalid credentials.")
        attempts += 1
```

### Example 6: Grade Book 📊

```python
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

top_student = max(grades, key=grades.get)
print("Top student:", top_student, "with grade", grades[top_student])
```

### Example 7: Shopping Cart 🛒

```python
cart = {}

while True:
    item = input("Add item (or 'quit'): ")
    if item == "quit":
        break
    if item in cart:
        cart[item] += 1
    else:
        cart[item] = 1

print("Final cart:", cart)
```

## 3) Sets vs Dictionaries

| Feature        | Set              | Dictionary                |
|----------------|-----------------|---------------------------|
| Data type      | Collection of unique values | Collection of key-value pairs |
| Access by      | Value membership | Keys                      |
| Example use    | Unique IDs, tags | Lookup table, mapping     |

## Wrap-Up
- **Sets**: unordered, unique elements, great for membership and set operations
- **Dictionaries**: key-value mappings, very fast lookups
- Both are hash-based, efficient but unordered
