## 1. Variable Assignment

- Analogy: Think of a box with a label; the label is the variable name, the content is the value;

- Assignment with "="

````python
user_name = 'Emma'
````

**Note**: Naming conventions in python
- **Snake Case**: Words are lowercase, separated by underscores "_"; commonly used in Python for variables and functions; 
- **Camel Case**: Words are joined without underscores; first word starts lowercase, following words start uppercase; common in JavaScript variables & methods;
- **Pascal Case** (sometimes called Upper Camel Case): Similar to Camel Case, but the first word is also capitalized; Common in Python class names and many other languages.

## 2. Data Types
### 2.1 Core Data types
**Integers (`int`)**: whole numbers

**Floats (`float`)**: decimal numbers

**Strings (`str`)**: text in quotes

**Booleans (`bool`)**: `True` or `False`

Use `type(x)` to check a variable's type.

````python
age = 20           # int (without decimal point)
pi = 3.14159       # float
name = 'Emma'      # str
is_student = True  # bool

print(age, type(age))
print(pi, type(pi))
print(name, type(name))
print(is_student, type(is_student))
````

### 2.2 Type Conversion
#### 2.2.1 Implicit Conversion (Type Promotion)
Python automatically converts smaller/less precise types to larger/more precise ones.

Example: int + float → result is float

````python
x = 10      # int
y = 3.5     # float
z = x + y   # int + float → float
print(z)
print(type(z))
````

#### 2.2.2 Explicit Conversion (Type Casting)
Use built-in functions to manually change types:

- `int()` → convert to integer (decimals truncated)
- `float()` → convert to float
- `str()` → convert to string
- `bool()` → convert to boolean

````python
# Convert string to int
num_str = '100'
num = int(num_str)
print(num + 50)   # 150

# Convert int to string
age = 21
msg = 'I am ' + str(age) + ' years old.'
print(msg)

# Bool conversion
print(bool(0))    # False
print(bool(5))    # True
print(bool(''))   # False
print(bool('Hi')) # True
````

## 3. Operators
### 3.1 Arithmetic Operators
`+`, `-`, `*`, `/` (true division), `//` (floor division), `%` (remainder), `**` (power)

````python
print('5 + 3 =', 5 + 3)
print('10 - 2 =', 10 - 2)
print('4 * 3 =', 4 * 3)
print('7 / 2 =', 7 / 2)
print('7 // 2 =', 7 // 2)
print('7 % 2 =', 7 % 2)
print('2 ** 3 =', 2 ** 3)
````

### 3.2 Comparison Operators (produce booleans)
`==`, `!=`, `>`, `<`, `>=`, `<=`

````python
print('5 > 3  ->', 5 > 3)
print('5 == 3 ->', 5 == 3)
print('5 != 3 ->', 5 != 3)
````

### 3.3 Logical Operators
`and`, `or`, `not`

````python
print('True and False ->', True and False)
print('True or False  ->', True or False)
print('not True       ->', not True)
````

### 3.4 Python math package for more operators
import the math package first

````python
import math

print(math.sqrt(16))     # 4.0
print(math.pow(2, 3))    # 8.0 (same as 2 ** 3)

print(math.sin(math.pi/2))  # 1.0
print(math.cos(0))          # 1.0
print(math.tan(math.pi/4))  # 1.0

print(math.factorial(5))    # 120
print(math.comb(5, 2))      # 10  (5 choose 2)
print(math.perm(5, 2))      # 20  (5 permute 2)
````
