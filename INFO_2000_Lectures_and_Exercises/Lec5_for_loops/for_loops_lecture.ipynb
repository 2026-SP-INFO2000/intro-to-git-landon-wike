### Recall the the number guessing game

Given a target value, input your guess, and determine the result of your guess based on the following conditions
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"

Run your program in the following code cell multiple times until you find the target value. You can use a counter to track how many guesses you make.

````python
import random
target = random.randint(1, 10)   # random int between 1 and 10 (inclusive)

max_num = 10 # maximum number of guesses
for num in range(max_num):
    guess = int(input("Please input your guess:"))

    if guess == target:
        print("Correct!")
        break
    elif guess == target - 1 or guess == target + 1:
        print(f"{guess} is close!")
    elif guess > target+1:
        print(f"{guess} is too High!")
    else:
        print(f"{guess} is too Low!")

    print(f"There are {max_num-num-1} guesses remaining!")
````

### 2) `range()` and Index‑based Loops
```
for index in range(start, stop):
    BODY
```

- `range(start, stop)` - `start`:`stop`-1 (`start` is inclusive, `stop` is exclusive)
- `range(stop)` - 0:`stop`-1
- `range(start, stop, step)` 

````python
print('range(5):', list(range(5)))
print('range(2, 7):', list(range(2, 7)))
print('range(0, 10, 2):', list(range(0, 10, 2)))
````

````python
fruits = ["apple", "banana", "cherry"]
for i_fruit in range(len(fruits)):
    print("I like", fruits[i_fruit])
````

````python
grades = [88, 92, 79, 93]
for i in range(len(grades)):
    print("Index:", i, "Score:", grades[i])
````

### 5) Nested loops
For loop inside another for loop

Print a **Pascal Triangle**
```
Row 0:        1
Row 1:       1 1
Row 2:      1 2 1
Row 3:     1 3 3 1
Row 4:    1 4 6 4 1
```


````python
# Pascal's Triangle with for loops

num_rows = 6   # number of rows

triangle = []

for i in range(num_rows):
    # start each row with all 1s
    row = [1] * (i + 1)

    # fill in the middle numbers
    for i_col in range(1, i):
        row[i_col] = triangle[-1][i_col-1] + triangle[i_row-1][i_col]

    triangle.append(row)

# print row by row
for row in triangle:
    print(row)
````

### 6) List Comprehension
Powerful **succint** syntax for creating lists based on any iterable

`RES = [EXPRESSION for VARS in ITERABLE]` is equivalent to 
```
RES = []
for VARS in ITERABLE:
    RES.append(EXPRESSION)
```



`RES = [EXPRESSION for VARS in ITERABLE if CONDITION]`
is equivalent to 
```
RES = []
for VARS in ITERABLE:
    if CONDITION:
        RES.append(EXPRESSION)
```


````python
# Loop version
squares = []
for n in range(5):
    squares.append(n**2)

# Comprehension version
squares = [n**2 for n in range(5)]
print(squares)  # [0, 1, 4, 9, 16]
````

````python
# Loop version
evens = []
for n in range(10):
    if n % 2 == 0:
        evens.append(n)

# Comprehension version
evens = [n for n in range(10) if n % 2 == 0]
print(evens)  # [0, 2, 4, 6, 8]
````

````python
# Loop version
pairs = []
for x in [1,2,3]:
    for y in ['a',b'']:
        pairs.append([x,y])
        
print(pairs)

# Comprehension version
pairs = [[x,y] for x in [1,2,3] for y in ['a','b']]
print(pairs)

````

### 7) More Examples

````python
# 1) Counting
grades = [88, 92, 79, 93]
count_A = 0
for g in grades:
    if g >= 90:
        count_A += 1
print('A grades:', count_A)
````

````python
# 2) Summing / Average
total = 0
for g in grades:
    total += g
    
avg = total / len(grades)
print('Average:', avg)
````

````python
# 3) Searching
fruits = ["apple", "banana", "cherry"]
found = False
for f in fruits:
    if f == "banana":
        found = True
        print('Found banana!')
        break
print('found =', found)
````


Given `text = "programming is fun"`, count how many vowels (a, e, i, o, u) appear.

````python
text = "programming is fun"
vowels = "aeiou"
count = 0
for ch in text:
    if ch in vowels:
        count += 1
print('Vowel count:', count)
````
