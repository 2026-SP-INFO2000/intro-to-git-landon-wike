### Exercise 2
Given a list of words, create a list of tuples `(word, length)` using for loop.

```python
words = ["python", "data", "science", "fun"]
# Your code here
results = []

for word in words:
    length = len(word)
    results.append((word,length))

print(results)
```

### Exercise 3
Check if a string is the same forward and backward (Palindrome). Use the input function to input a string.

```python
string = "hannah"

is_palindrome = True
for i in range(len(string)//2):
    if string[i] != string[-(i+1)]:
        is_palindrome = False
        break
print("Palindrome?", is_palindrome)
```

### Exercise 4 Password Strength Checker 
Check if a password has at least 1 digit, 1 uppercase, 1 lowercase letter, and 1 special character from `!@#$%^&*`

```python
pwd = "PyTh0n123"  # or you can use the input function to input your own passwords
has_digit = False
has_upper = False
has_lower = False
has_spec = False
spec_char = '!@#$%^&*'
for ch in pwd:
    if ch.isdigit(): 
        has_digit = True
    if ch.isupper(): 
        has_upper = True
    if ch.islower():
        has_lower = True
    if ch in spec_char:
        has_spec = True

print("Does password have requirements?", has_digit and has_upper and has_lower and has_spec)
```
