## 1) Tuples 
- Container type that is very similar to a list, except it is **immutable**, meaning it cannot change in length or contents once it is created
- **Immutable** objects are stored and accessed slightly more **efficiently** than lists
- Created by a comma-separated list of objects optionally enclosed by parentheses
`(obj0,obj1,obj2,…,objK)` or
`obj0,obj1,obj2,…objK`



### 1.1) Indexing and slicing

```python
numbers = 1, 2, 3
print(numbers)
print(numbers[0], numbers[-1])     # indexing
print(numbers[1:3])          # slicing

# Immutability
# numbers[0] = 99            # ❌ not allowed
```

### 1.2) Tuple Unpacking
You can assign a tuple of variables to a tuple of objects


```python
coords = (5, 10)
x, y = coords
print(x, y)
```

Swap the values in a tuple `(a, b)`.

```python
pair = (10, 20)
a, b = pair
pair = (b, a)
print(pair)
```

### Tuples in Practice
- Tuples are used to return multiple values from functions


```python
def min_max(nums):
    return min(nums), max(nums)

lo, hi = min_max([2, 7, 1, 8])
print(type(min_max([2, 7, 1, 8])))
print(lo, hi)
```

### 1.3) zipping into tuples
`zip(LIST1,LIST2,…,LISTK) -> zip`
Result is a zip iterator, which can yield successive tuples formed by the elements at aligning offsets in each list/tuple
list(zip(A,B)) -> [(A[0],B[0]),(A[1],B[1]),…,(A[K],B[K])]


```python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

zipped = zip(names, ages)
print(type(zipped)) # a zip object is not a list — it’s an iterator, which means you can loop over it once, or convert it to a container
#print(list(zipped))
print(tuple(zipped))
# [('Alice', 25), ('Bob', 30), ('Charlie', 35)]
```

```python
# If the lists are not the same length, zip() stops at the shortest one:
names = ["Alice", "Bob"]
ages = [25, 30, 35]

print(list(zip(names, ages)))
```

- Iterating with for loops

```python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

for name, age in zip(names, ages):
    print(f"{name} is {age} years old")
```

**Iterator and iterables**
- An iterator is an object that represents a stream of data; it remembers its current position and can return the next item when asked.
- An iterable is something you can loop over (lists, strings, tuples, sets)
- All iterators are iterables (because they have `__iter__()`).
- Not all iterables are iterators.
For example, a list is iterable, but not an iterator itself. You can get an iterator from it with `iter(list)`.
- for loops in Python work by calling iter() on the iterable, then repeatedly calling next() on the iterator behind the scenes.

- Unzipping with `*`

```python
pairs = [("Alice", 25), ("Bob", 30), ("Charlie", 35)]
names, ages = zip(*pairs)
print(type(names)) # The output is a tuple
print(names)  # ('Alice', 'Bob', 'Charlie')
print(ages)   # (25, 30, 35)
```

## 2) Strings
Strings are **tuples** of (Unicode) characters, hence are **immutable** once created


```python
str_test = "Emma's phone"
str_test
```

### 2.1) String Literals
- Single-quotes are useful if the string includes double-quotes
- Double-quotes are useful if the string includes single-quotes
- Multi-line strings, which are enclosed by 3 single-quotes or 3 double-quotes, are useful for including new lines
- Backslashes followed by a character, C, i.e., \C are a way of directly including non-typeable or otherwise invalid characters
- These are called escape sequences
- Therefore, “\” is not possible by itself, and instead is “\\”


```python
print("\\")
print(r"Hello\nWorld")   # newline
print("Hello\tWorld")   # tab
print("C:\\new_folder") # literal backslash
```

### 2.2) Raw and formatted strings
- If you may have a lot of escape characters, you may choose to use a raw string, which is **prefixed with an r**, and ignores escape sequences
- If you want to include a variable in a string, you can use a formatted string, which is **prefixed with an f**, and includes variables in curly braces
- F-Strings and R-Strings can be combined with the **rf prefix**



```python
s = r"What is a\a?"
print(s)
name = "Emma"
age = 8
print(f"My name is {name}, and I am {age} years old")
```

### 2.3) Indexing and slicing

```python
s = "python"
print(s[0], s[-1])   # indexing
print(s[1:4])        # slicing
for ch in s:
    print(ch, end=" ")
```

```python
s = "python"
s += '#'
print(s)

```

```python
s = "hello"
# s[0] = "H"    # ❌ TypeError
print(s.upper())     # returns new string
```

### 2.3) Common string operations
- `len(STRING) -> int`
Returns the number of characters in STRING
- `STRING.upper() and STRING.lower() -> str`
Return STRING that is all upper or all lower case
- `STRING.strip() -> str`
Return STRING that has leading and trailing whitespace removed
- `STRING.join(ITERABLE) -> str`
Return all ITERABLE elements as a string separated by STRING.
Note that ITERABLE must be strings
- `STRING.split(SEPARATOR) -> list`
Split STRING into individual list elements based on SEPARATOR
- `STRING.replace(STRING1,STRING2) -> str`
Return a string with every instance of STRING1 in STRING replaced by STRING2

```python
msg = "  data-science  "
print(msg.upper())
print(msg.strip())       # trim spaces
print(msg.replace("data", "info"))
print(msg.split("-"))
print(" ".join(["AI","ML"]))
```

```python
str1 = '-'.join([str(i) for i in range(10)])
print(str1)
```

```python
dir(str) # show all the operations of strings
```

### 2.4) Comparing Strings
- Characters are numbers, so they can be numerically compared
(See the underlying number with `ord(CHARACTER)`)
- Lowercase letter is larger than uppercase letter
- `STRING1 == STRING2`
`True` if the length of the strings match, and every character is the same underlying value
- `STRING1 > STRING2` or `STRING < STRING2`
Compared at the first character that does not match


```python
ord('a')
```

```python
str1 = 'I love math'
str2 = 'I love Math'
print(str1 > str2)
```

```python
sentence = "python is fun"
# a) Uppercase
print(sentence.upper())

# b) Split into words
sentence_list = sentence.split(' ')
print(sentence_list)

# c) Replace 'fun' with 'powerful'
sentence_list[-1] = 'powerful'
print(sentence_list)
```

```python
line = "alice,20,biology"

# a) Split into parts
line_list = line.split(',')
print(line_list)

# b) Convert age to int
line_list[1] = int(line_list[1])
print(line_list)

# c) Store as tuple (name, age, major)
line_tuple = tuple(line_list)
print(line_tuple)
```

## Examples

### Example 1
Given a list of point tuples `(x, y)`, print those that lie on the diagonal (`x == y`).

```python
points = [(1,2), (3,3), (5,7), (4,4)]

for x, y in points:
    if x == y:
        print((x,y))
```

### Example 2
Replace vowels in a string with `*`.

```python
msg = "programming is fun"
vowels = "aeiou"
coded = ""
for ch in msg:
    if ch in vowels:
        coded += "*"
    else:
        coded += ch
print(coded)
```

### Example 3  Palindrome Checker 
Check if a string is the same forward and backward.


```python
s = "racecar"
is_palindrome = True
for i in range(len(s)//2):
    if s[i] != s[-(i+1)]:
        is_palindrome = False
        break
print("Palindrome?", is_palindrome)
```

### Example 4 Most Frequent Letter 
Find which letter appears most often in a string.

```python
word = "banana"
max_count = 0
most_letter = ''
for ch in word:
    count = word.count(ch)
    if count > max_count:
        max_count = count
        most_letter = ch
print(most_letter, "appears", max_count, "times")
```

### Example 5 Password Strength Checker 
Check if a password has at least 1 digit, 1 uppercase, and 1 lowercase letter.

```python
pwd = "PyTh0n123"
has_digit = False
has_upper = False
has_lower = False

for ch in pwd:
    if ch.isdigit(): # check if the current character is a digit
        has_digit = True
    if ch.isupper(): # check if the current character is uppercase
        has_upper = True
    if ch.islower(): # check if the current character is lowercase
        has_lower = True

print("Strong password?", has_digit and has_upper and has_lower)
```
