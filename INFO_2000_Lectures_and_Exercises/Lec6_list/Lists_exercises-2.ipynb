### Exercise 2
1. Create `a = [1,2,3]` and set `b = a`. Change `a[0] = 99`.
What is b now? (print it) Why?
2. Properly copy `a` into `c` so that changing `a` doesn't change `c`.
3. Let `d = [a]*3` and `d[0][0] = 100`, print `d`. How to make the element in the 1st row and first column change?

```python
a = [1,2,3]
b = a
a[0] = 99
print(b)
# b = a does not create a new list, so when you change a, you're also changing the same list memory for b
```

```python
a = [1,2,3]
c = a.copy()

#checking it works properly
a[0] = 99
print(a)
print(c)
```

```python
d = [a]*3
d[0][0] = 100
print(d)

#To make the element in the 1st row and column change you have to create indepedent copies of a 

a = [1,2,3]
d = [a.copy() for _ in range(3)]

d[0][0] = 100
print(d)
```

### Exercise 3
Given `playlist = ["Bluebird", "Sunrise", "Neon Lights", "Echoes"]`
Perform the following, using only indexing/slicing/methods and control flow

1. Swap the first and last songs.
2. Insert "Interlude" as the second song.
3. Replace the third and fourth songs with ["Remix A", "Remix B"] using slice assignment.
4. If "Sunrise" is still in the playlist, move it to the end (remove then append).
5. Print the final playlist.

```python
playlist = ["Bluebird", "Sunrise", "Neon Lights", "Echoes"]
## Add your code in the following

playlist[0],playlist[-1] = playlist[-1], playlist[0]

playlist.insert(1, "Interlude")

playlist[2:4] = ["Remix A", "Remix B"]

if "Sunrise" in playlist:
    playlist.remove("Sunrise")
    playlist.append("Sunrise")

print(playlist)

```

### Exercise 4
Import the random library using `import random`
1. Create a list of 10 types of pizza toppings and use the `random.sample()` function to randomly select 3 types to order as a pizza tonight. Print the pizza toppings.

```python
import random 

pizza_toppings = ["Pepperoni", "Sausage", "Jalopenos", "Green Peppers", "Onions", "Olives", "Bacon", "Spinach", "Pineapple", "Ham"]

pizza_order = random.sample(pizza_toppings, 3)

print('The toppings for tonights pizza will be:', pizza_order)
```

2. Create a list of 5 local restaurants with their pricing (e.g. last resort grill: \$\$\$, Cali n Tito's: \$\$, restautrant3: \$, etc), and use the `random.shuffle()` function to randomly select one to go to for dinner tonight. Print the selected restaurant with a helpful message.

```python
restaurants = ["Taco Mama:$", "Baddies Burgers:$", "Ahi:$$", "Sully Steamers:$", "Last Resort Grill:$$$"]

random.shuffle(restaurants)

selected_resturant = restaurants[0]

print("Our randomly selected restaurant is:", selected_resturant)
```

3. Create a list of 10 outdoor activities (e.g., soccer, hiking, biking, kayaking), and use the `random.choice()` function to randomly select one activity to do this weekend.

```python
outdoor_activies = ["Soccer", "Hiking", "Swimming", "Biking", "Disc Golf", "Kayaking", "Tennis", "Softball", "Baseball", "Camping"]

selected_activity = random.choice(outdoor_activies)

print("The activity we will be doing this weekend is:", selected_activity)
```

### Exercise 5
You are given three items in your backpack: `bag = ["map", "flashlight", "snack"]`.
1. Ask the user (with input) if it's night ("y" / "n"). If night, ensure "flashlight" is in the bag (add it only if missing).
2. Ask if they're hungry. If yes, remove "snack" if present.
3. Print the bag.

```python
bag = ["map","flashlight","snack"]

if_night = input(str('Is it night time? (y/n)'))

if if_night == "y":
    if "flashlight" not in bag:
        bag.append("flashlight")
print(bag)

if_hungry = input(str('Are you hungry? (y/n)'))

if if_hungry == "y":
    if "snack" in bag:
        bag.remove("snack")

print(bag)
```
