### 1) What is a List?
- A **list** is a **container** type that can group **any** types of object together in numbered sequence.
- Lists are mutable (can be modified and extended) and indexed (retrieved by numbered position).
- Python has **List**, **Set**, **Tuple**, and **Dictionary** types built-in, which have special operators for creating them and manipulating them
- Created using square brackets `[]`.

```python
numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]
mixed = [1, "hello", 3.5, True]
print(numbers)
print(fruits)
print(mixed) # print multiple lists at the same time
```

### 2) Indexing & Slicing
- Zero-based indexing
- Negative indices count from the end
- Slicing: taking a portion (a slice) of a list by specifying where to start, where to stop, and (optionally) the step size.
syntax `list[start:stop:step]`

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
first = fruits[0] # the first index is 0
last = fruits[-1] # -1 refers to the last index
second_from_end = fruits[-2] #
slice1 = fruits[1:4] # start index is inclusive, end index is exclusive 
slice2 = fruits[:3] # from start to index 2
slice3 = fruits[::2]
slice4 = fruits[::-2] # negative step size refers to going backwards
first, last, second_from_end, slice1, slice2, slice3, slice4
```

### 3) Modifying Lists
- Replace values
- Add with `append`, `insert`
- Remove with `remove`, `pop`, `del`

```python
colors = ["red", "green", "blue"]
colors[1] = "yellow"  # re-assignment to modify entry 1
print(type(colors))
print(colors)
colors.append("purple")
print(colors)
colors.insert(0, "black")
print(colors)
colors.remove("blue")
print(colors)
popped = colors.pop()
del colors[2]
colors, popped
```

#### **Note**:  The difference between `append` and `extend`
- `List1.append(Obj)` adds `Obj` to the end of the list; `Obj` can be any data type;
- `List1.extend(List2)` merges elements of `List2` to that in `List1`; equivalent to `List1 += List2` or `List1 = List1 + List2`

```python
List_1 = [1, 2, 'a', True]
List_2 = List_1[:] # rerun with List_2 = List_1
List_1.append(['c','d'])
print(List_1)
print(List_2)
List_2.extend('10')
print(List_2)
id(List_1), id(List_2)
```

### 4) List copying
For a list `a`, after assignment `b = a`, `b` is **not** a copy of `a`, just another name or alias of `a`
- The reason is that a list name is the **address/pointer** for the stored data in memory; thus, `a` and `b` have the same address; can check using the `id()` function

```python
a = [1,2,3]
b = a
a[0] = 99
print(a)
print(b)
id(a), id(b)
```

- Ways to make a new list with the same elements

```python
a = [1, 2, 3]

b = a.copy()      # method
print(b)
c = list(a)       # constructor
print(c)
d = a[:]          # slicing
print(d)

id(a), id(b), id(c), id(d)
```

- The above does not work for **nested lists**

```python
a = [[1,2], [3,4]]
b = a.copy()
c = a[:]

a[0][0] = 99
print(a)   # [[99, 2], [3, 4]]
print(b)   # [[99, 2], [3, 4]]  ← inner list changed too!
print(c)
print(id(a))
print(id(b))
print(id(c))
id(a[0]), id(b[0]), id(c[0])
```

- Use list comprehension (will introduce after introducing for loops) to avoid this issue or deep copy

```python
a = [[1,2], [3,4]]
b = [row[:] for row in a]
print(b)
a[0][0] = 99
a, b
```

- Alternatively, use deep copy from the copy package

```python
import copy

a = [[1,2], [3,4]]
b = copy.deepcopy(a)

a[0][0] = 99
print(a)   # [[99, 2], [3, 4]]
print(b)   # [[1, 2], [3, 4]]  ← completely independent
```

### 5) List multiplication
for a `List`, `List*3` is the same as `List = List + List + List` or extend your list

```python
List = [1, 2, 100]
List*3
```

```python
a = [[0]*3]*3
print(a)
a[0][0] = 99
print(a)
```

```python
a = [[0] * 3 for _ in range(3)]
print(a)  
a[0][0] = 99
print(a)
```

### 6) Useful Functions & Methods
- Python built-in functions: `sum(List)`, `max(List)`, `min(List)`, `sorted(List)`
- `List.count()`, `List.index()`, `List.reverse()`, `List.sort()`
- `List.sort()` vs. `sorted(List)`

```python
nums = [3, 6, 1, 9, 2, 6]
length = len(nums)
total = sum(nums)
mn, mx = min(nums), max(nums)
print('Minimum number is:', mn)
print('maximum number is', mx)
count6 = nums.count(6)
idx9 = nums.index(9)
nums_sorted = sorted(nums)
nums.sort() # in-place sort
nums.reverse()
length, total, mn, mx, count6, idx9, nums_sorted, nums
```

### 7) the **in** operator for searching in a list
syntax: `res = target **in** List`
- res is **True** if target is in List, otherwise **False**
- Note the **in** keyword will be used later in **for** loop, but for a different purpose.

```python
students = ["Ana", "Bo", "Chen", "Drew"]
for s in students:
    print("Hello,", s)
```

### 8) Nested Lists
Lists of lists (2‑D data, like matrices)

```python
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
grid

print(grid[:2][:2])
print(grid[:2]) # slices rows(outer list)
```

- Indexing Rows and Columns

```python
sub_grid = [row[:2] for row in grid[:2]]
sub_grid
```

```python
center = grid[1][1]
last_col = [row[-1] for row in grid]
center, last_col
```
