## Exercise 1 — File Reader
Write a function `read_file(filename)` that:
- Reads the file and prints it if it exists
- Prints `'File not found'` if missing

```python
def read_file(filename):
    # put your code here
    try:
        with open(filename, 'r') as file:
            # We use .read() to get the content and return it
            return file.read()
    except FileNotFoundError:
        return 'File not found'

# Create a sample file and test
with open("sample.txt", "w") as f:
    f.write("hello world")

print(read_file("sample.txt"))     # contents
print(read_file("missing.txt"))    # File not found
```

## Exercise 2: Password Validator

Write a function `set_password(pw)` that:
- Raises a `ValueError` if the password length is less than 6.
- Raises a `ValueError` if the password does not contain at least one digit (Hint: use `any(ch.isdigit() for ch in pw)`).
- Otherwise, returns `"Password set successfully"`.

In the main program, ask the user for a password inside a `try/except` block and print either the success message or the error message.

```python
def set_password(pw):
    # put your code here
# Rule 1: Length check
    if len(pw) < 6:
        raise ValueError("Password must be at least 6 characters long.")
    
    # Rule 2: Digit check
    if not any(ch.isdigit() for ch in pw):
        raise ValueError("Password must contain at least one digit.")
    
    return "Password set successfully"

# main program starts here
user_input = input("Enter a new password: ")

try:
    result = set_password(user_input)
    print(result)
except ValueError as e:
    # 'e' captures the specific message we wrote in the 'raise' statements
    print(f"Error: {e}")
```

## Exercise 3 — Number Guessing Game with Exception Handling
- Generate a random number (1–10).
- Ask the user to guess.
- Use `try/except` to catch non-integer guesses.
- Give feedback (`Too high!`, `Too low!`).

```python
import random

def guessing_game():
    # 1. Generate a random number (1–10)
    secret_number = random.randint(1, 10)
    print("Guess the number 1 - 10")

    while True:
        # 2. Ask the user to guess
        user_input = input("Enter your guess: ")

        # 3. Use try/except to catch non-integer guesses
        try:
            guess = int(user_input)
        except ValueError:
            print("Invalid input! Please enter a whole number.")
            continue  # Skip the rest of the loop and ask again

        # 4. Give feedback (Too high!, Too low!)
        if guess < secret_number:
            print("Too low!")
        elif guess > secret_number:
            print("Too high!")
        else:
            print("Congratulations! You guessed it.")
            break  # Exit the loop since they won

# Start the game
guessing_game()
```
