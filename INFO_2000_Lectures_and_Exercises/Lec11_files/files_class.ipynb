## 1) Opening a File in Python
- Basic operation
` f = open(FILEPATH: str, MODE: str = “r”)`
  -  Opens the file specified by FILEPATH, which can be relative or absolute in specified MODE, returns a **file object** called `f`

| Mode | Meaning |
|------|----------|
| `"r"` | Open for **reading** (default). File must already exist. |
| `"w"` | Open for **writing**. If file exists, it is **truncated (cleared)** first. If it doesn’t exist, a new file is created. |
| `"x"` | Open for **exclusive creation**. Fails if the file already exists. |
| `"a"` | Open for **appending**. New data is written at the end of the file (keeps old data). Creates the file if it doesn’t exist. |
| `"b"` | **Binary mode** (used with other modes). Example: `"rb"` or `"wb"`. |
| `"t"` | **Text mode** (default). Example: `"rt"` (same as `"r"`). |
| `"+"` | Open for **updating** (reading **and** writing). Example: `"r+"`, `"w+"`. |

- Files must be opened prior to performing operations

**Common file operations**
- Result depends on the file mode
- Reading
  - `file.read()` -> all text/bytes of the file
  - `file.read(N)` -> N characters/bytes of the file 
  - `file.readline()` -> One line string of a text file
  - `file.readlines()` -> List of lines
- Writing
file.write(contents: str) -> writes the contents to the file
file.flush() -> forces any “buffered” write operations to be completed


````python
f = open('example.txt','w')
f.write('The first line') # overwrite or create a new file
#f.close()
# For two operations in two modes, the file has to be closed and reopened
f = open('example.txt','r')
content = f.read()
print(content)
f.close()
````

````python
f = open('example.txt','w')
for i in range(5):
    f.write(f'\nLine {i+1}')
f.close()

f = open('example.txt','r')
line1 = f.readline()
line2 = f.readline()
f.close()
print(line1)
print(line2)
````

````python
conent = f.readline()
print(type(content))
````

**Note:** Although not always strictly necessary, it is highly recommended to close a file (`file.close()`) before the next operation, because
- Free resources → Open files consume system resources (file descriptors). Too many open files can cause errors.
- Flush data → If you’re writing, data may sit in a buffer. Closing ensures all data is actually written to disk.
- Avoid conflicts → Another process (or even your own program) may need the file. Leaving it open could cause locking issues.

## 2) With Statement 
```
with OPENSTATEMENT as VAR:
    STATEMENTBODY
```
**Automatically** opens and closes files used by subsequent code


- Opening and Reading Files

````python
# Demo: create a sample file, then read it entirely
sample_text = """Roses are red
Violets are blue
Python writes files
And so can you
"""
# Write the content in sample_text to example.txt
with open("example.txt", "w", encoding="utf-8") as f:
    f.write(sample_text)
## This is equivalent to
# f = open("example.txt", "w", encoding="utf-8")
# f.write(sample_text)
# f.close()

# Read the content in example.txt
with open("example.txt", "r", encoding="utf-8") as f:
    content = f.read()
print(content)
````

````python
# Reading line by line with a for-loop
with open("example.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.rstrip())  # rstrip to remove trailing newline from the right
````

**Other read helpers:**

- `f.readline()` → one line at a time
- `f.readlines()` → list of all lines (watch memory on big files)


- Writing to Files

````python
# Overwrite vs Append
with open("output.txt", "w") as f:
    f.write("First line\n")
    f.write("Second line\n")

with open("output.txt", "a") as f:  # append
    f.write("Appended line\n")

print(open("output.txt", "r").read())
````

- Processing Files

````python
# Count words in a file
with open("example.txt", "r", encoding="utf-8") as f:
    words = f.read().split()
print("Word count:", len(words))
````

## Example 1 — Write a Poem ✍️
Create a file `poem.txt` and write three lines into it using `with open(..., "w")`.

````python
lines = [
    "Sunlight on the window",
    "Coffee in a warm mug",
    "Coding in the morning",
]
with open("poem.txt", "w", encoding="utf-8") as f:
    for line in lines:
        f.write(line + "\n")
print("poem.txt created.")
````

## Example 2 — Read Your Poem 📖
Read `poem.txt` and print it line by line using a `for` loop.

````python
with open("poem.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        print(f"{i:>2}: {line.rstrip()}")
        # right-aligns the line number in a 2-character space
````

## Example 3 — Word Counter 📊
Write a program that counts how many words are in `poem.txt`.

````python
with open("poem.txt", "r", encoding="utf-8") as f:
    text = f.read()
word_count = len(text.split())
print("Total words:", word_count)
````

## Example 4 — Longest Word Finder 🏆
Find the longest word in `poem.txt`.


````python
def longest_word_from_file(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        words = f.read().split()
    # Tie-breaker: keep the first longest word found
    longest = ""
    for w in words:
        if len(w) > len(longest):
            longest = w
    return longest

print("Longest word:", longest_word_from_file("poem.txt"))
````

## Example 5 — Shopping List 🛒
Write a function that:
1) Saves a list of grocery items into `shopping.txt`.
2) Reads the file and prints the items with line numbers.

````python
def save_shopping(items, path="shopping.txt"):
    with open(path, "w", encoding="utf-8") as f:
        for item in items:
            f.write(str(item) + "\n")

def print_shopping(path="shopping.txt"):
    with open(path, "r", encoding="utf-8") as f:
        for i, line in enumerate(f, start=1):
            print(f"{i}: {line.rstrip()}")

save_shopping(["apples", "bread", "milk"]) 
print_shopping()
````

## 3) The `os` Module

The `os` module lets us interact with the operating system.  
It’s useful for tasks like:
- Checking the current working directory
- Listing files/folders
- Creating or removing directories
- Joining file paths safely


````python
import os

# Get current working directory
print("Current directory:", os.getcwd())

# List files in the current directory
print("Files:", os.listdir())
print(type(os.listdir())) # The output type is a list

# Make a new folder (won’t error if it already exists)
os.makedirs("test_folder", exist_ok=True)
print("After creating folder:", os.listdir())

# Change into the new folder and show where we are
os.chdir("test_folder")
print("Now inside:", os.getcwd())

# Go back up one level
os.chdir("..")
print("Back to:", os.getcwd())
````

### Key Functions to Remember:
- `os.getcwd()` → current working directory
- `os.listdir(path)` → list files and folders
- `os.makedirs(path, exist_ok=True)` → create a folder (safe if it exists)
- `os.chdir(path)` → change working directory
- `os.remove(file)` → delete a file
- `os.rmdir(folder)` → delete an empty folder
