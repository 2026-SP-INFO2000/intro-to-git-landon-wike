## Exercise 1 — Joke Writer 🤡
- Create a file called `jokes.txt`;

- Write 3 jokes into it (each on a new line);

- Then reopen it, read all lines, and print them with line numbers.

```python
with open("jokes.txt", "w") as file:
    file.write("Why do programmers prefer dark mode? Because light attracts bugs.\n")
    file.write("How many Gophers does it take to change a lightbulb? Just one, but it's a concurrent lightbulb.\n")
    file.write("Why was the function so happy? Because it had no arguments.\n")

with open("jokes.txt", "r") as file:
    for index, line in enumerate(file, start=1):
        print(f"{index}. {line.strip()}")

# Expected output example:
# 1: Why did the Python cross the road? To get to the other IDE!
# 2: ...
# 3: ...
```

## Exercise 2 — File Extension Sorter 📂
- In the current directory, create a few dummy files: `a.txt, b.py, c.md, d.py`;

- Write a program that:

 - 1. Scans the current directory using os.listdir().

 - 2. Groups files into a dictionary by their file extension.

```python
import os

files_to_create = ['a.txt', 'b.py', 'c.md', 'd.py']
for f in files_to_create:
    open(f, 'a').close()

file_groups = {}


for filename in os.listdir('.'):
   
    if filename in files_to_create:
       
        name, ext = os.path.splitext(filename)
        
        if ext not in file_groups:
            file_groups[ext] = []
            
        file_groups[ext].append(filename)

print(file_groups)

# Expected output example:
# {'.txt': ['a.txt'], '.py': ['b.py', 'd.py'], '.md': ['c.md']}
```

## Exercise 3 — Copycat 🐱
Write a program that:

- 1. Creates a text file `story.txt` with a short story;

- 2. Reads it and creates a copy called `story_copy.txt`;

- 3. Print confirmation: `"Copied story.txt to story_copy.txt"`.

```python

story_content = """Once upon a time, a small cat named Pixel 
found a way to jump inside a computer screen. 
Now, he lives in the cloud, chasing digital mice."""

with open("story.txt", "w") as original:
    original.write(story_content)

with open("story.txt", "r") as source:
    data = source.read() 
    
with open("story_copy.txt", "w") as destination:
    destination.write(data)

print(f"Copied story.txt to story_copy.txt")
```
