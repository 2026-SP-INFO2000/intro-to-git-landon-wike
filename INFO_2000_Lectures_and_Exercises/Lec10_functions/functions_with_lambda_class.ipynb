# Functions in Python

**Learning goals:**
- Understand what a function is and why it’s useful
- Learn the syntax: `def name(parameters):`
- Use parameters and return values

## 0) Motivation
- Functions allow you to reuse code by making it vary according to parameters
- Once a function is created and tested, it can often be used without knowing the details of how it works 


## 1) Function Basics
```python
def FUNCTIONNAME(INPUTVAR1, INPUTVAR2): # input variables are optional
    FUNCTIONBODY
    return FUNCTIONRESULT # This is also optional
```

````python
def greet():
    print("Hello, world!")

greet()
````

## 2) Parameters & Arguments

````python
def greet(name):
    print("Hello,", name)

greet("Alice")
greet("Bob")
````

## 3) Return Values

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
````

## 4) Functions with default values
```python
def FUNCTIONNAME(INPUTVAR1, INPUTVAR2=VALUE):
    FUNCTIONBODY
    return FUNCTIONRESULT
```

This function can be called as 
```
    FUNCTIONNAME(VALUE1) or
    FUNCTIONNAME(VALUE1,VALUE2)
```


````python
def power(base, exponent=2):
    return base ** exponent

print(power(5))        # 25 (uses default exponent=2)
print(power(5, 3))     # 125 (uses given exponent=3)
````

## 5) Functions with type hints
`def FUNCTIONNAME(INPUTVAR:TYPE) -> RETURNTYPE:`

- TYPE is the expected type of INPUTVAR, e.g. str, int, float, list
- RETURNTYPE is the expected return type of FUNCTIONNAME, e.g. str, int, float list
- Type hints allow you “hint” at the type, even though it is not actually enforced


````python
def add(x: int, y: int) -> int:
    return x + y

print(add(3, 5))   
print(add(3.0, 5.0)) 
````

## 6) Local vs Global Variables

````python
x = 10

def func():
    x = 5
    print("Inside:", x)

func()
print("Outside:", x)
````

````python
x = 10

def func():
    global x
    x = 5   # modifies global variable
    print("Inside function:", x)

func()
print("Outside function:", x)   # now 5
````

## 7) Multiple Return Values

````python
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([3, 7, 2, 9])
print("Min:", low, "Max:", high)
````

### Example 1: Dice Roller 🎲

````python
import random

def roll_dice():
    return random.randint(1, 6)

print("You rolled:", roll_dice())
````

### Example 2: Temperature Converter 🌡️

````python
def c_to_f(c):
    return (c * 9/5) + 32

def f_to_c(f):
    return (f - 32) * 5/9

print("20°C =", c_to_f(20), "°F")
````

### Example 3: Simple Calculator ➕➖✖️➗

````python
def calculator(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        return a / b
    else:
        return "Invalid operator"

print(calculator(5, 3, "+"))
````

### Example 4: Random Password Generator 🔑

````python
import string

def generate_password(length):
    chars = string.ascii_letters + string.digits
    return "".join(random.choice(chars) for _ in range(length))

print("Password:", generate_password(8))
````

### Example 5: Guess the Number Game 🎯

````python
def guess_number():
    number = random.randint(1, 10)
    guess = 0
    while guess != number:
        guess = int(input("Guess (1-10): "))
        if guess < number:
            print("Too low!")
        elif guess > number:
            print("Too high!")
    print("Correct! The number was", number)

# guess_number()  # Uncomment to play interactively
````

### Example 6: Shopping Cart with Functions 🛒

````python
cart = {}

def add_item(item):
    if item in cart:
        cart[item] += 1
    else:
        cart[item] = 1

def show_cart():
    return cart

add_item("apple")
add_item("banana")
add_item("apple")
print(show_cart())
````

### Example 7: Pig Latin Translator 🐷

````python
def pig_latin(word):
    return word[1:] + word[0] + "ay"

print(pig_latin("python"))
````

## 8) Lambda Functions

Lambda functions are short, anonymous functions written in a single line.

**Syntax:**
```python
F = lambda VARS: EXPRESSION
```
Equivalent to
```python
def anonymous(VARS):
     return EXPRESSION
F = anonymous
```
- Term comes from a branch of mathematics called Lambda Calculus
- VARS are a tuple of input variable names
- EXPRESSION is the function body
- F is the function object
- Like regular functions, except
  - Lambda functions can be assigned when they are created
  - Lambda functions are anonymous (no variable name)
  - Lambda functions only have one expression inside them


### Example 8 — Simple Addition

````python
add = lambda a, b: a + b
print(add(3, 5))
````

### Example 9 — Square a Number

````python
square = lambda x: x**2
print(square(4))
````

### Example 10 — Sorting with Lambda

````python
names = ["Alice", "Bob", "Charlie", "David"]
names.sort(key=lambda x: len(x))
print(names)
````

### Example 11 — Filtering with Lambda

````python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
````

### Example 12 — Mapping with Lambda

````python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
# The map() function applies a given function to each item in an iterable 
# (like a list, tuple, or string) and returns an iterator
print(squares)
````

An equivalent way

````python
def square(x):
    return x * x

nums = [1, 2, 3, 4]
squares = list(map(square, nums))
print(squares)
````
