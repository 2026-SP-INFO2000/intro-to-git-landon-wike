- Ask for the first name using the input() function
- Ask for the last name using the input() function
- Print out your full name using the print() function
- Show the number of characters in your first name using the len() function
- Show the number of characters in your last name using the len() function


```python
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
print("Your full name is:", first_name, last_name)
print("Number of characters in your first name:", len(first_name))
print("Number of characters in your last name:", len(last_name))
```
