1. Given `text = "programming is fun"`, use a dictionary to show how many times each of vowels `(a, e, i, o, u)` appear in the text, print the dictionary. (5 points)

````python
text = "programming is fun"

vowels = "aeiou"

vowel_counts = {v: 0 for v in vowels}

for char in text.lower():
    if char in vowel_counts:
        vowel_counts[char] += 1

print(vowel_counts)
````

2. You have a dictionary representing a store’s inventory:
```python
inventory = {"apple": 5, "banana": 2, "orange": 0}
```
- Ask the user what they want to buy.
- If the item is available (quantity > 0), reduce the count by 1 and confirm purchase.
- If not, print `'Sorry, out of stock'`.
- Stop when the user types `'quit'`.  (5 points)

````python
inventory = {"apple": 5, "banana": 2, "orange": 0}

while True:
    item = input("What would you like to buy? (type 'quit' to exit): ").lower()
    
    if item == 'quit':
        print("Goodbye!")
        break
    
   
    if item in inventory:
       
        if inventory[item] > 0:
            inventory[item] -= 1 
            print(f"You purchased an {item}. Confirming purchase!")
        else:
            print("Sorry, out of stock")
    else:
        print("We don't have that item in our store.")
````

4. Password Checker
- Write a function `check_password(pwd)`to check if a password is valid. A valid password should have least 8 characters, and has at least 1 digit, 1 uppercase, 1 lowercase letter, and 1 special character from `!@#$%^&*`.

- In the main program, use the input function to input the password and call the function `check_password(password)` to check the password. (5 points)

````python
def check_password(pwd):
    
    if len(pwd) < 8:
        return False
    
    has_digit = False
    has_upper = False
    has_lower = False
    has_special = False
    special_chars = "!@#$%^&*"
    
    
    for char in pwd:
        if char.isdigit():
            has_digit = True
        elif char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char in special_chars:
            has_special = True

    return has_digit and has_upper and has_lower and has_special


password_input = input("Please enter a password to check: ")

if check_password(password_input):
    print("Valid password")
else:
    print("Invalid password. Ensure it has 8+ characters, a digit, uppercase, lowercase, and a special character.")
````

5. Given the maximum number of guesses and a target value (randomly generated between 1 and 10), use for loop or while loop to simulate the number guessing game. In each loop, input your guess and determine the result of your guess based on the following conditions
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low" 
- Use formatted string to print how many guesses remaining at the end of each loop. (5points)

````python
import random
target = random.randint(1,10)
max_guesses = 10 
count = 0 
while count < max_guesses:
    count += 1
    guess = int(input('Enter your guess: '))

    if guess == target:
        print("Correct!")
        break
    elif guess == target -1 or guess == target +1:
        print("Close!")
    elif guess > target +1:
        print("Too High")
    else:
        print("Too Low")


    guess_remian = {max_guesses - count}
    print('You have', guess_remian,'guesses remaining')

if count == max_guesses:
    print('You are out of guesses')


````

6. Write a program to print Pascal triangle **upside down** for a given number of rows. (6 points)

 A Pascal triangle is
```
Row 0:    1
Row 1:    1 1
Row 2:    1 2 1
Row 3:    1 3 3 1
Row 4:    1 4 6 4 1
```

````python
def print_upside_down_pascal(n):
    triangle = []

    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
        triangle.append(row)

    for i in range(n - 1, -1, -1):
       
        
        for val in triangle[i]:
            print(val, end=" ")
        print()


rows = int(input("Enter the number of rows: "))
print_upside_down_pascal(rows)
````

**Bonus** Print the centered Pascal Triangle (3 points)
```
      1
     1 1
    1 2 1
   1 3 3 1
  1 4 6 4 1
 1 5 10 10 5 1
```

````python
def print_centered_pascal(n):
    triangle = []

    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
        triangle.append(row)

    for i in range(n):
        print(" " * (n - i), end="")
        
        for val in triangle[i]:
            print(val, end=" ")
        print()


rows = int(input("Enter the number of rows: "))
print_centered_pascal(rows)
````
