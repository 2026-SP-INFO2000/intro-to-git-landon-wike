## A motivating example (Exercise 4 from last class)

Write a program that asks the user for the coefficients (a,b, and c) of a quadratic equation and computes and prints out the **real** root (x) using the following formula:
$ ax^2+bx+c=0 \implies x=\frac{-b \pm \sqrt{b^2-4ac}}{2a}$

````python
# get the inputs from the user:
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

if b**2 - 4*a*c < 0:
    print("b^2 should be larger than 4ac! Consider increasing the value of b!")
else:
    numerator1 = -b + (b**2 - (4*a*c))**0.5
    numerator2 = -b - (b**2 - (4*a*c))**0.5
    solution1 = numerator1/(2 *a)
    solution2 = numerator2/(2 *a)
    print(f"x = {solution1} or {solution2}")
````

**Motivation**: Programs need conditional statements because input is not known at the time a program is written

### Syntax
```
if CONDITION1:
    BODY1
elif CONDITION2:
    BODY2
elif CONDITION3:
    BODY3
else:
    BODYELSE
```

- **If**, **elif**, and **else** are keywords;  **:** delineates the start of a BODY
- elif and else are optional parts
- BODY1,2,3,ELSE is a set of statements all indented by the same amount of whitespace 
- CONDITION1,2,3 are expressions (potentially complex ones) that are evaluated to (or convert to) True or False
- Only the first CONDITION that evaluates to True will have its BODY executed
- If none of the CONDITIONS are True, the BODYELSE will execute
- Conditions usually given by comparison operators (produce booleans) `==`, `!=`, `>`, `<`, `>=`, `<=`

### Nested Conditionals & Boolean Logic

**Concepts:**
- Combining conditions with `and`, `or`, `not` (always use parentheses with compound statements because precedence is much less likely to be clear)
- Nested conditionals (if inside if)

### Example 4
 Ask user for age and whether they have an ID (True/False). Print if they can vote.

````python
age = float(input("What is your age?"))
has_id = bool(input("Do you have an ID? Enter 1 for yes, 0 for no"))
if age >= 18 and has_id:
    print('You can vote!')
else:
    print('Sorry! You cannot vote!')
````

````python
# Another way using nested conditionals
age = float(input("What is your age?"))
has_id = bool(input("Do you have an ID? Enter 1 for yes, 0 for no"))
if age >= 18:
    if has_id:
        print('You can vote!')
    else:
        print('You cannot vote!')
else:
    print('You cannot vote!')
````

### Example 8
Write a program that classifies a triangle by side lengths (equilateral, isosceles, scalene).

````python
a = int(input('Enter side a: '))
b = int(input('Enter side b: '))
c = int(input('Enter side c: '))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print('Equilateral triangle')
    elif a == b or b == c or a == c:
        print('Isosceles triangle')
    else:
        print('Scalene triangle')
else:
    print('Not a valid triangle')
````

### Example 9 Shipping Cost by Weight

Rules:

If weight ≤ 1 kg → $5

If 1 < weight ≤ 3 kg → $8

If 3 < weight ≤ 5 kg → $12

If 5 < weight ≤ 10 kg → $20

If weight > 10 kg → $35

````python
weight = float(input("Enter package weight (kg): "))

if weight <= 1:
    cost = 5
elif weight <= 3:
    cost = 8
elif weight <= 5:
    cost = 12
elif weight <= 10:
    cost = 20
else:
    cost = 35

print(f"Shipping cost: ${cost}")
````
