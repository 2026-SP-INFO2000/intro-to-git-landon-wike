## Exercise 1 – Leap Year Checker
Determine if a given year is a leap year.
- A year is a leap year if divisible by 4 but not 100, unless also divisible by 400.

```python
year = int(input('Enter year: '))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, 'is a leap year!')
else:
    print(year, 'is not a leap year')
```

## Exercise 2– ATM Withdrawal
Given a balance in your bank account, input an withdrawal amount, check if it is a valid withdrawal, and return the remaining balance for a valid withdrawal

```python
balance = 4000
withdrawal = float(input('Enter withdrawl amount: '))

if withdrawal < balance:
    remaining = balance - withdrawal
    formatted_remaining = f"{remaining:.2f}"
    print('Your remaining balance is $',formatted_remaining)
else: 
    print('Invalid withdrawal')
```

## Exercise 4 – Grading with Plus/Minus
 Input a score and return the corresponding grade based on the following 
 - A+: 97-100
 - A: 93-96
 - A-: 90-92
 - B+: 87-89
 - B: 83-86
 - B-: 80-82
 - C+: 77-79
 - C: 73-76
 - C-: 70-72
 - D: 60-69
 - F: <60 

```python
grade = float(input('Enter your grade: '))

if grade >= 97:
    print('A+')
elif grade >= 93:
    print('A')
elif grade >= 90:
    print('A-')
elif grade >= 87:
    print('B+')
elif grade >= 83:
    print('B')
elif grade >= 80:
    print('B-')
elif grade >= 77:
    print('C+')
elif grade >= 73:
    print('C')
elif grade >= 70:
    print('C-')
elif grade >= 60:
    print('D')
else:
    print('F')
```

### Exercise 5
