### The `in` operator in loops and conditionals

````python
text = "I like python"
vowels = "aeiou"
cnt = 0

for ch in text:
    if ch in vowels:
        cnt += 1

print(f"There are {cnt} vowels in the text")
````

````python
lst_1 = [1, 2, 4, 5]
lst_2 = [2,4,5,6]
common = []

for item in lst_1:
    if item in lst_2:
        common.append(item)

print(common)
````
